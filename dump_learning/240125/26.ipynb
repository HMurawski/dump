```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Symulacja danych sprzedaży
np.random.seed(42)
data = np.random.randint(50, 500, size=(12, 5))

# Etykiety produktów i miesięcy
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Tworzenie DataFrame
sales_df = pd.DataFrame(data, columns=products, index=months)
print(sales_df)
```

```python
#2.
suma_kazdy_produkt = sales_df.sum(axis=0)  # Sumowanie kolumn
print(suma_kazdy_produkt)

```

```python
srednia_mies_produkty = sales_df.mean(axis=0)
srednia_mies_produkty
```

```python
total_sum = 0
for index, value in enumerate(suma_kazdy_produkt):
    total_sum += value
print(total_sum)
```

```python
total_sum = sales_df.values.sum()
print(total_sum)
```

```python
udzial = round(suma_kazdy_produkt / total_sum * 100,2)
print(udzial)
```

```python
fig, ax = plt.subplots()
ax.pie(udzial, radius=3, center=(4,4),wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True, labels=udzial.index, autopct="%1.1f%%")
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))
ax.legend(udzial,
          title="Udzial Produktow",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))
ax.set(xlim=(0, 8), ylim=(0, 8))
ax.axis("off")
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

# Wykres słupkowy
sales_df.plot(kind='bar', ax=ax, width=0.8, colormap='viridis')

# Dostosowanie wykresu
ax.set_title("Trendy miesięczne sprzedaży", fontsize=16)
ax.set_xlabel("Miesiące", fontsize=12)
ax.set_ylabel("Sprzedaż", fontsize=12)
ax.legend(title="Produkty", loc="upper right", fontsize=10)
ax.grid(axis='y', linestyle="--", alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

# Suma sprzedaży w czasie
suma_miesieczna = sales_df.sum(axis=1)
ax.plot(suma_miesieczna, marker="o", color="blue", linewidth=2, label="Suma sprzedaży")

# Dostosowanie wykresu
ax.set_title("Suma sprzedaży w czasie", fontsize=16)
ax.set_xlabel("Miesiące", fontsize=12)
ax.set_ylabel("Sprzedaż", fontsize=12)
ax.legend(loc="upper left", fontsize=10)
ax.grid(visible=True, linestyle="--", alpha=0.7)

# Dodanie etykiet dla punktów
for i, value in enumerate(suma_miesieczna):
    ax.text(i, value + 50, f"{value:.0f}", ha="center", fontsize=10)

plt.tight_layout()
plt.show()
```
