```python
# Dodaj nową kolumnę Port Classification:

# "Deep Water" dla portów z Channel Depth (m) większą niż 15.
# "Moderate Water" dla portów z Channel Depth (m) między 10 a 15.
# "Shallow Water" dla portów z Channel Depth (m) mniejszą niż 10.

def get_depth(depth):
    if depth > 15:
        return "Deep Water"
    elif depth >=10 and depth <=15:
        return "Moderate Water"
    else:
        return "Shallow Water"

df["Port Classification"] = df["Channel Depth (m)"].apply(get_depth)
```

```python
import numpy as np

conditions = [
    df["Channel Depth (m)"] > 15,
    (df["Channel Depth (m)"] >= 10) & (df["Channel Depth (m)"] <= 15),
    df["Channel Depth (m)"] < 10
]
choices = ["Deep Water", "Moderate Water", "Shallow Water"]
df["Port Classification"] = np.select(conditions, choices)
```

```python
#Dodaj nową kolumnę Coordinates, która połączy Latitude i Longitude w jeden string w formacie "(lat, lon)".
def get_coords(row):
    if not pd.isnull(row["Latitude"]) and not pd.isnull(row["Longitude"]):
        return f"({row['Latitude']}, {row['Longitude']})"
    else:
        return None

df["Coords"] = df.apply(get_coords, axis=1)
```

```python
#"Country Code"
def get_port_type(code):
    if code == "United States":
        return "Domestic"
    else:
        return "International"

#two solutions - one with function, other with lambda    
df["Port Type"] = df["Country Code"].apply(lambda x: "Domestic" if x=="United States" else "International")
```
