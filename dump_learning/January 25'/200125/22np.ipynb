```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arr2d = np.random.randint(low=1,high=101, size=(5,5))
print(f"max w kazdej kolumnie: {np.max(arr2d, axis=0)}")
print(f"min w kazdym wierszu: {np.min(arr2d, axis=1)}")
```

```python
mean_arr2d = np.mean(arr2d)
std_arr2d = np.std(arr2d)

standarized_arr2d = (arr2d - mean_arr2d) / std_arr2d
print(standarized_arr2d)
```

```python
#przekatna tablicy
przekatna = np.diag(arr2d)
sum_of_diagonal = np.sum(przekatna)
print(sum_of_diagonal)
```

```python
weather_data = {
    "Temp" : np.random.randint(low=-10, high=36, size=(365)),
    "Opady mm" : np.random.randint(low=0, high=50, size=(365)),
    "Day" : np.arange(1,366)
}
df_weather = pd.DataFrame(weather_data)
```

```python
df_weather.head(3)
```

```python
avg_temp = weather_data["Temp"].mean()
max_temp = weather_data["Temp"].max()
suma_opadow = np.sum(df_weather["Opady mm"])
hot_days = len(df_weather[df_weather["Temp"] > 30])
print(f"Srednia dziennia temperatura: {round(avg_temp,2)}")
print(f"Max dziennia temperatura: {round(max_temp,2)}")
print(f"Suma opadow: {suma_opadow} mm.")
print(f"Liczba goracych dni: {hot_days}.")
```

```python
df_temp = df_weather["Temp"]
plt.figure(figsize=(14,8))
plt.plot(df_weather["Day"], df_weather["Temp"], label='Temp', color='blue')

hot_days_idx = df_weather["Temp"] > 30
plt.scatter(
    df_weather["Day"][hot_days_idx], 
    df_weather["Temp"][hot_days_idx], 
    color="red", 
    label="Dni z temp. > 30C", 
    zorder=5)

plt.xlabel("Day of Year")
plt.ylabel("Temperature")
plt.title("Temp over Year")
plt.axhline(avg_temp, color="green", linestyle="--", label=f"rednia temp.: {round(avg_temp, 2)}C")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```
