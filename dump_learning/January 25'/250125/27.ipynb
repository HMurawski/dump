```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
s_h_than50 = matrix[matrix > 50].sum()
print(s_h_than50)
s_wierszy = matrix.mean(axis=1)
print(s_wierszy)
s_kolumn = matrix.mean(axis=0)
print(s_kolumn)
```

```python
unikalne = np.unique(matrix)
print(unikalne)
print(f"liczba unikalnych wartosci: {len(unikalne)}")
```

```python
values, counts = np.unique(matrix, return_counts=True)
print("Wartości i ich liczność:")
for val, count in zip(values, counts):
    print(f"{val}: {count}")
```

```python
print(f"Mediana: {np.median(matrix)}")
print(f"Minimum: {matrix.min()}")
print(f"Maksimum: {matrix.max()}")
```

```python
np.random.seed(42)

# Generowanie danych
normal_data = np.random.normal(0, 1, 10_000)  # Rozkład normalny
uniform_data = np.random.uniform(0, 1, 10_000)  # Rozkład jednostajny
exponential_data = np.random.exponential(1.0, 10_000)  # Rozkład wykładniczy
binomial_data = np.random.binomial(20, 0.5, 10_000)  # Rozkład dwumianowy
```

```python

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Rozkład normalny
axs[0, 0].hist(normal_data, bins=50, color='blue', alpha=0.7)
axs[0, 0].set_title("Rozkład normalny")
axs[0, 0].set_xlabel("Wartości")
axs[0, 0].set_ylabel("Liczność")
axs[0, 0].grid(visible=True, linestyle="--", alpha=0.5)

# Rozkład jednostajny
axs[0, 1].hist(uniform_data, bins=50, color='green', alpha=0.7)
axs[0, 1].set_title("Rozkład jednostajny")
axs[0, 1].set_xlabel("Wartości")
axs[0, 1].set_ylabel("Liczność")
axs[0, 1].grid(visible=True, linestyle="--", alpha=0.5)

# Rozkład wykładniczy
axs[1, 0].hist(exponential_data, bins=50, color='orange', alpha=0.7)
axs[1, 0].set_title("Rozkład wykładniczy")
axs[1, 0].set_xlabel("Wartości")
axs[1, 0].set_ylabel("Liczność")
axs[1, 0].grid(visible=True, linestyle="--", alpha=0.5)

# Rozkład dwumianowy
axs[1, 1].hist(binomial_data, bins=50, color='purple', alpha=0.7)
axs[1, 1].set_title("Rozkład dwumianowy")
axs[1, 1].set_xlabel("Wartości")
axs[1, 1].set_ylabel("Liczność")
axs[1, 1].grid(visible=True, linestyle="--", alpha=0.5)

# Globalny tytuł i układ
fig.suptitle("Porównanie różnych rozkładów (10,000 punktów)", fontsize=16)
plt.tight_layout()
plt.subplots_adjust(top=0.92)
plt.show()
```

```python
months = [f"{year}-{month:02d}" for year in range(2021, 2024) for month in range(1, 13)]
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
data = np.random.randint(50, 1001, size=(len(months), len(products)))


sales_data = pd.DataFrame(data, columns=products, index=months)
sales_data.index.name = "Month"

print(sales_data.head())
```

```python
trend_miesieczny = sales_data.sum(axis=1)
best_produkt = sales_data.sum(axis=0).sort_values(ascending=False).index[0]
worst_produkt = sales_data.sum(axis=0).sort_values(ascending=False).index[-1]
print(f"{best_produkt}\n{worst_produkt}")
```

```python
total_sales = sales_data.values.sum()
sales_by_product = sales_data.sum(axis=0)

pct_per_product = round(sales_by_product / total_sales *100,2)
pct_per_product
```

```python
plt.figure(figsize=(12,6))
plt.plot(trend_miesieczny, linestyle="--", marker="o", color="g", linewidth=2)
plt.title("Suma sprzedaży w czasie", fontsize=16)
plt.xlabel("Miesiące", fontsize=12)
plt.ylabel("Sprzedaż", fontsize=12)
plt.xticks(rotation = 90)

for i, value in enumerate(trend_miesieczny):
    plt.text(i, value + 50, f"{value:.0f}", ha="center", fontsize=10)
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.pie(pct_per_product, radius=3, center=(4,4),frame=True, labels=pct_per_product.index, autopct="%1.1f%%")
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))
ax.legend(pct_per_product,
          title="Udzial Produktow",
          bbox_to_anchor=(1, 0, 0.5, 1))
ax.set(xlim=(0, 8), ylim=(0, 8))
ax.axis("off")
```
