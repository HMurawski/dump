```python
import pandas as pd
```

```python
df = pd.read_csv('customers_extended.csv')
df.head()
```

```python
# Oblicz średnią kwotę wydaną przez klienta w każdym regionie.
# Znajdź klientów, którzy wydali więcej niż średnia dla ich regionu.
# Dodaj kolumnę PurchaseRank, która przypisuje rangę klientom w każdym regionie na podstawie TotalSpent (od najwyższej do najniższej).
# Wyszukaj klientów, którzy są w top 10% pod względem wydatków (TotalSpent) w swoim regionie.
```

```python
## Oblicz średnią kwotę wydaną przez klienta w każdym regionie.
avg_spent_per_klient_region = df.groupby(["Region", "Name"])['TotalSpent'].mean()
df_avg_spent_per_klient_region = avg_spent_per_klient_region.reset_index()
df_avg_spent_per_klient_region
```

```python
## Znajdź klientów, którzy wydali więcej niż średnia dla ich regionu.
avg_per_region = df.groupby("Region")['TotalSpent'].mean()
df_avg_region = avg_per_region.reset_index()
#df.rename(columns={"A": "a"
df_avg_region = df_avg_region.rename(columns={"TotalSpent": "Avg_Spent_Per_Region"})
df_avg_region

#merging both dfs
df_merged = df.merge(df_avg_region, on="Region", how='left')
#difference between total spent and spent per region
#df_merged['TotalSpent'] - df_merged["Avg_Spent_Per_Region"]
df_merged["Spent_More_Than_Avg"] = df_merged["TotalSpent"] > df_merged["Avg_Spent_Per_Region"]
df_merged
```

```python
df['Avg_Spent_Per_Region'] = df.groupby("Region")['TotalSpent'].transform('mean')
df["Spent_More_Than_Avg"] = df["TotalSpent"] > df["Avg_Spent_Per_Region"]
df
```

```python
## Dodaj kolumnę PurchaseRank, która przypisuje rangę klientom w każdym regionie na podstawie TotalSpent (od najwyższej do najniższej).
df['PurchaseRank'] = df.groupby('Region')['TotalSpent'].rank(method='max', ascending=False)
df
```

```python
# Wyszukaj klientów, którzy są w top 10% pod względem wydatków (TotalSpent) w swoim regionie.
#df['Region'].value_counts()
import numpy as np
def get_top_10_pct(group):
    group = group.sort_values(by="TotalSpent", ascending=False)
    n = len(group)
    top_n = max(1, int(np.ceil(0.1 * n)))
    return group.iloc[:top_n]  

top10 = df.groupby("Region", group_keys=False).apply(get_top_10_pct)



print(top10)    
```

```python
# 3. Łączenie datasetów
# Dane:
# Połącz dane z sales_extended.csv i customers_extended.csv.

# Zadania:
# Znajdź klientów, którzy zrobili zakupy w więcej niż jednym regionie.
# Oblicz całkowitą wartość transakcji (Amount) przypisaną do każdego klienta w DataFrame.
# Znajdź region z największą średnią wartością transakcji przypisaną do klientów.
df_sales = pd.read_csv('sales_extended.csv')
df_merged = df.merge(df_sales, on='Region', how='inner')
df_merged
```

```python
# Znajdź klientów, którzy zrobili zakupy w więcej niż jednym regionie.
regions_per_customer = df_merged.groupby("Name")["Region"].nunique()
customer_with_multi_region = regions_per_customer[regions_per_customer > 1].index
df_check = df_merged[df_merged["Name"].isin(customer_with_multi_region)]
df_check #none of the client bought items in more than 1 region
```

```python
# Oblicz całkowitą wartość transakcji (Amount) przypisaną do każdego klienta w DataFrame.
df_merged.groupby("Name")["Amount"].sum().sort_values(ascending=False)
```

```python
# Znajdź region z największą średnią wartością transakcji przypisaną do klientów.
avg_per_reg = df_merged.groupby("Region")["Amount"].mean().sort_values(ascending=False).reset_index()
#avg_per_reg
top_region = avg_per_reg.iloc[0]
top_region
```
