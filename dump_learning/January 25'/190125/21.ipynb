```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(997)
data_stack = {
    "Country": np.random.choice(["USA", "Germany", "Japan"], 100),
    "Year": np.random.choice([2020, 2021],100),
    "Population": np.random.randint(low=1000, high=100000, size=(100)),
    "GDP": np.random.randint(low=50, high=5000, size=(100))
}
df_stack = pd.DataFrame(data_stack)
```

```python
df_stack.head()
```

```python
#Obliczyć średnią populację i sumę PKB (GDP) dla każdego kraju
mean_pop_sum_gdp = round(df_stack.groupby(["Country"])[["Population","GDP"]].agg(["mean", "sum"]),2)
mean_pop_sum_gdp
```

```python
#Policzy różnicę między maksymalną a minimalną wartością Population
def get_pop(group):
    return group["Population"].max() - group["Population"].min()

pop_range = df_stack.groupby("Country").apply(get_pop, include_groups=False)
pop_range.reset_index(name="Population Range")
```

```python
pop_range_agg = df_stack.groupby("Country")["Population"].agg(lambda x: x.max() - x.min())
pop_range_agg = pop_range_agg.reset_index(name="Population Range")
print(pop_range_agg)
```

```python
df_stack[df_stack["Country"] == "Japan"]["Population"].min()
```

```python
data_pivot = {
    "Country": ["USA", "USA", "Germany", "Germany", "Japan", "Japan"],
    "Year": [2020, 2021, 2020, 2021, 2020, 2021],
    "Product": ["A", "B", "A", "B", "A", "B"],
    "Sales": [500, 700, 300, 400, 200, 300]
}
df_pivot = pd.DataFrame(data_pivot)
data_stack = {
    "Country": ["USA", "USA", "Germany", "Germany", "Japan", "Japan"],
    "Year": [2020, 2021, 2020, 2021, 2020, 2021],
    "Population": [331, 332, 83, 84, 126, 125],
    "GDP": [21137, 22940, 3845, 4223, 5065, 5315]
}
df_stack = pd.DataFrame(data_stack)
```

```python
df = df_stack.merge(df_pivot, on=["Country", "Year"])
df
```

```python
sum_gdp_sales = df.groupby("Country")[["Sales", "GDP"]].sum()
sales_per_gdp = round(sum_gdp_sales["Sales"] / sum_gdp_sales["GDP"],2)
df_sales = sales_per_gdp.reset_index()
df = df.merge(df_sales, on="Country")
```

```python
df = df.rename(columns={"0_x":"Sales per GDP"})
df.head()
```

```python
df_pivot_multi = df_pivot.set_index(["Country", "Year"])
df_pivot_multi
```

```python
sales_sum = df_pivot_multi.groupby(level="Country")["Sales"].sum()
sales_sum
```

```python
sales_sum_reset = sales_sum.reset_index()
sales_sum_reset.rename(columns={"Sales": "Total Sales"}, inplace=True)


print(sales_sum_reset)
```
