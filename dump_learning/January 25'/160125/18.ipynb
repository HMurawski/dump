```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

data_stack = {
    "Country": ["USA", "USA", "Germany", "Germany", "Japan", "Japan"],
    "Year": [2020, 2021, 2020, 2021, 2020, 2021],
    "Population": [331, 332, 83, 84, 126, 125],
    "GDP": [21137, 22940, 3845, 4223, 5065, 5315]
}
df_stack = pd.DataFrame(data_stack)


data_pivot = {
    "Country": ["USA", "USA", "Germany", "Germany", "Japan", "Japan"],
    "Year": [2020, 2021, 2020, 2021, 2020, 2021],
    "Product": ["A", "B", "A", "B", "A", "B"],
    "Sales": [500, 700, 300, 400, 200, 300]
}
df_pivot = pd.DataFrame(data_pivot)


data_multiindex = {
    "Region": ["North America", "North America", "Europe", "Europe", "Asia", "Asia"],
    "Product": ["A", "B", "A", "B", "A", "B"],
    "Sales": [500, 700, 300, 400, 200, 300],
    "Profit": [200, 300, 100, 150, 50, 75]
}
df_multiindex = pd.DataFrame(data_multiindex)
```

```python
ax = df_pivot.groupby(["Year", "Country"])["Sales"].sum().plot(kind='barh', title="Sales in 2020/2021")
ax.set_xlabel("Sales")
ax.set_ylabel("Year, Country")
```

```python
df_multiindex.groupby("Region")["Sales"].sum().plot(
    kind='pie', autopct='%1.1f%%', title="Sales by Region")
```

```python
#uklad liniowych rownan
# Rozwiązanie układu równań
# 2x + y = 8
# x - 3y = -11
A = np.array([[2, 1], [1, -3]])  # Macierz współczynników
b = np.array([8, -11])           # Wektor wyników

# Rozwiązanie
solution = np.linalg.solve(A, b)
print("Rozwiązanie układu równań:", solution)
```

```python
np.random.seed(42)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
stores = ['Store_A', 'Store_B', 'Store_C', 'Store_D', 'Store_E']

data = pd.DataFrame(
    np.random.randint(2000, 5000, size=(6, 5)),
    index=months,
    columns=stores
)
print(data)
```

```python
data.plot.line(title="Monthly Sales Trends", legend=False, figsize=(8,6), linewidth = 2, marker="o")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(visible=True, linestyle="--", alpha=0.7)
```

```python
data.sum().plot(kind="bar", color=["skyblue", "orange", "green", "red", "purple"], figsize=(6,4))
plt.xticks(rotation=45)
plt.xlabel("Stores")
plt.ylabel("Sales")
```
