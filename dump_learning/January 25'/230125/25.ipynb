```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tab = np.random.randint(1,101, 1000)
```

```python
plt.figure(figsize=(10,6))
plt.hist(tab, bins=10, color='green', alpha=0.7)
plt.title("Rozklad liczb")
plt.xlabel("Liczba")
plt.ylabel("Ilosc wystapien")
plt.grid(visible=True, linestyle="--", alpha=0.5)
```

```python
counts, bins, patches = plt.hist(tab, bins=10, color='green', alpha=0.7)
for count, bin in zip(counts, bins):
    plt.text(bin + 5, count + 2, str(int(count)), ha='center')
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].hist(tab, bins=5, color='blue', alpha=0.7)
axs[0].set_title("Histogram z 5 przedziaami")
axs[0].set_xlabel("Liczba")
axs[0].set_ylabel("Ilo wystpie")
axs[0].grid(visible=True)

axs[1].hist(tab, bins=20, color='red', alpha=0.7)
axs[1].set_title("Histogram z 20 przedziaami")
axs[1].set_xlabel("Liczba")
axs[1].set_ylabel("Ilo wystpie")
axs[1].grid(visible=True)

plt.tight_layout()
plt.show()
```

```python
x = np.arange(0,10.1, 0.1)
f_sin = np.sin(x)
f_cos = np.cos(x)
```

```python

fig, ax = plt.subplots(figsize=(8,6))
ax.plot(x, f_sin, linewidth=2.0, color="c", label="Sinus", linestyle="-.", marker="o")
ax.plot(x, f_cos, linewidth=2.0, color="k", label="Cosinus", linestyle="--", marker="s")
ax.set_xlabel("X values")
ax.set_ylabel("Y values")
ax.grid(visible=True, linestyle="--", alpha=0.5)
ax.axhline(0, color="gray", linewidth=1.0, linestyle="--")
ax.legend()
fig.suptitle("Sinus vs Cosinus")

```

```python
np.random.seed(42)
X = np.random.normal(loc=0, scale=1, size=(500))
epsilon = np.random.normal(loc=0, scale=0.5, size=500) 
Y = 2 * X + epsilon  # Zaleno Y = 2X + epsilon
```

```python

fig, ax = plt.subplots(figsize=(10,6))
sizes = np.abs(Y) * 50 
colors = np.random.uniform(15, 150, len(X))
scatter = ax.scatter(X, Y, s=sizes, c=colors, cmap="viridis", vmin=0, vmax=150)
ax.grid(True, alpha=0.7)
fig.suptitle("Scatter")
fig.tight_layout()
ax.set_xlabel("X values")
ax.set_ylabel("Y values")
fig.colorbar(scatter, ax=ax, label="Intensity")
ax.plot(X, 2 * X, color="red", linewidth=2, label="Regression line")
ax.legend()

```
