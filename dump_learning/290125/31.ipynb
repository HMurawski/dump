```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)

# Zadanie 1: Dane sprzedaży dla 12 miesięcy
months = [f"2023-{month:02d}" for month in range(1, 13)]
sales = np.random.randint(1000, 5000, size=12)
df_sales = pd.DataFrame({"Month": months, "Sales": sales})

# Zadanie 2: Dane sprzedaży dla 5 produktów
products = ["Product A", "Product B", "Product C", "Product D", "Product E"]
product_sales = np.random.randint(5000, 20000, size=5)
df_product_sales = pd.DataFrame({"Product": products, "Sales": product_sales})

# Zadanie 3: Dane cen produktów (100 wartości)
prices = np.random.randint(50, 1000, size=100)
df_prices = pd.DataFrame({"Price": prices})

# Zadanie 4: Dane dla wykresu punktowego (X i Y)
X = np.random.randint(1, 100, size=100)
Y = X ** 2 + np.random.normal(0, 500, size=100)  # Kwadrat X + szum
df_scatter = pd.DataFrame({"X": X, "Y": Y})

# Zadanie 5: Dane dla wykresu kołowego (udział sprzedaży)
sales_pie = np.random.randint(10000, 50000, size=5)
df_pie = pd.DataFrame({"Product": products, "Sales": sales_pie})
```

```python
#z1
df_sales.plot(figsize=(10,6), marker="o", linestyle="--")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Sales over Months")

```

```python
fig, ax = plt.subplots(figsize=(10,6))
x = df_product_sales["Product"]
y = df_product_sales["Sales"]
ax.bar(x,y, width=1, edgecolor='white', linewidth=0.5, alpha=0.7)
ax.grid(True)

```

```python
plt.figure(figsize=(10,6))
sns.histplot(bins=10, data=(df_prices), kde=True, x="Price")
```

```python
sizes = np.random.uniform(50, 500, len(X))
colors = np.random.uniform(15, 80, len(X))
fig, ax = plt.subplots(figsize=(10,6))

ax.scatter(X, Y, s=sizes, c=colors, vmin=0, vmax=100)
ax.grid(True, alpha=0.7)
```

```python
labels = df_pie["Product"]
sizes = df_pie["Sales"]

fig, ax = plt.subplots(figsize=(8,6))
ax.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['olivedrab', 'rosybrown', 'gray', 'saddlebrown', "blue"], shadow={'ox': -0.04, 'edgecolor': 'none', 'shade': 0.9})
ax.legend(labels,loc="upper left",
          bbox_to_anchor=(1, 0, 0.5, 1), title="Products")
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
df_scatter["X"] = X
df_scatter["Y"] = Y


axs[0, 0].hist(bins=10, data=(df_prices),x="Price")
axs[0, 0].set_title("Rozkład normalny")
axs[0, 0].set_xlabel("Wartości")
axs[0, 0].set_ylabel("Liczność")
axs[0, 0].grid(visible=True, linestyle="--", alpha=0.5)


axs[0, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['olivedrab', 'rosybrown', 'gray', 'saddlebrown', "blue"], shadow={'ox': -0.04, 'edgecolor': 'none', 'shade': 0.9})



axs[1, 0].scatter(X, Y)
axs[1, 0].set_title("Scatter plot")
axs[1, 0].grid(visible=True, linestyle="--", alpha=0.5)



axs[1, 1].bar(x,y, width=1, edgecolor='white', linewidth=0.5, alpha=0.7)
axs[1, 1].set_title("Bar plot")
axs[1, 1].grid(visible=True, linestyle="--", alpha=0.5)
```
