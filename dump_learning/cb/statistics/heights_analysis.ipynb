```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("heights_with_outliers.csv")
df.head()
```

```python
df.shape
```

```python
df_mean = df["height"].mean()
df_std = df["height"].std()
print(f"mean: {df_mean}, std: {df_std}")
```

```python
sns.histplot(df.height, kde=True)

plt.axvline(df_mean, color="k", linestyle="dashed", linewidth=1)

#one std
plt.axvline(df_mean + df_std, color="r", linestyle="dashed", linewidth=1)
plt.axvline(df_mean - df_std, color="r", linestyle="dashed", linewidth=1)

#two std
plt.axvline(df_mean + 2*df_std, color="g", linestyle="dashed", linewidth=1)
plt.axvline(df_mean - 2*df_std, color="g", linestyle="dashed", linewidth=1)

#3 std
plt.axvline(df_mean + 3*df_std, color="b", linestyle="dashed", linewidth=1)
plt.axvline(df_mean - 3*df_std, color="b", linestyle="dashed", linewidth=1)
```
