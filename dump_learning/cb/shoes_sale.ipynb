```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("shoe_sales.csv")
df.head()
```

```python
len(df["date"])
```

```python
df.describe()
```

```python
df.info()
```

```python
nike_sales = df[df["brand"] == "Nike"]
nike_total = nike_sales["sold_qty"].sum()
nike_avg = nike_sales["sold_qty"].mean()
nike_median = nike_sales["sold_qty"].median()

print(f"Nike total sales: {nike_total}\nNike avg sales: {round(nike_avg,2)}\nNike median: {nike_median}")
```

```python
adidas_sales = df[df['brand'] == "Adidas"]
adidas_sales.shape
```

```python
adidas_sales.describe()
```

```python
adidas_sales.isnull().sum()
```

```python
adidas_sales.sold_qty.quantile([0.95])
```

```python
adidas_sales.plot(kind='line')
```

```python
adidas_sales = adidas_sales[adidas_sales["sold_qty"] < 20]
adidas_sales.describe()
```

```python

adidas_total = adidas_sales["sold_qty"].sum()
adidas_median = adidas_sales["sold_qty"].median()

print(f"Adidas total sales: {adidas_total}\nAdidas median: {adidas_median}")
```

```python
def plot_qty():
    plt.figure(figsize=(15,6))
    
    dates = nike_sales["date"]
    
    plt.plot(dates, nike_sales["sold_qty"], marker='o', label='Nike', color='blue')
    plt.plot(dates, adidas_sales["sold_qty"], marker='o', label='Adidas', color='green')
    
    plt.xlabel("Date")
    plt.ylabel("Total Qty Sold")
    plt.title("Daily Sales Qty for Nike and Adidas in September 2023")
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.show()
    
```

```python
plot_qty()
```
