```python
result = 0
tic = time.perf_counter()
for value in arr:
    result += value
toc = time.perf_counter()
print(f"suma: {result}")
print(f"Time using numpy {toc - tic:0.4f} seconds")
```

```python
import pandas as pd
np.random.seed(42)
data = {
    "Sold Units": np.random.randint(50, 500, size=12),
    "PPU": np.random.randint(10, 101, size=12),
    "Month": range(1, 13)
}
```

```python
df = pd.DataFrame(data)
print(df.head(3))
df["Rev"] = df["Sold Units"] * df["PPU"]
```

```python
# Miesiąc z największym przychodem
top_month = df.loc[df["Rev"].idxmax()]
print("\nMiesiąc z największym przychodem:")
print(top_month)
```

```python
# Średni miesięczny przychód
avg_revenue = df["Rev"].mean()
print("\nŚredni miesięczny przychód:", avg_revenue)
```

```python
# Liczba miesięcy z przychodem powyżej średniej
above_avg = df[df["Rev"] > avg_revenue]
print(above_avg)
```

```python
np.random.seed(42)
data = pd.DataFrame({
    "Store": np.random.choice(["Store_A", "Store_B", "Store_C", "Store_D"], size=24),
    "Month": np.tile(["Jan", "Feb", "Mar", "Apr", "May", "Jun"], 4),
    "Sales": np.random.randint(2000, 5000, size=24),
    "Region": np.random.choice(["North", "South", "East", "West"], size=24),
})
```

```python
store_a = data[data["Store"] == "Store_A"]
store_b = data[data["Store"] == "Store_B"]
store_c = data[data["Store"] == "Store_C"]

#sns.histplot()
```

```python
plt.figure(figsize=(10, 6))
for store in data["Store"].unique():
    store_data = data[data["Store"] == store]["Sales"]
    sns.histplot(
        store_data,
        bins=10,
        alpha=0.6,
        kde=True,
        label=store
    )
    
plt.title("Comparison of Sales Distributions by Store", fontsize=16)
plt.xlabel("Sales", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.legend(title="Stores")
plt.grid(visible=True, linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.boxplot(
    data=data,
    x="Store",
    y="Sales",
    palette='viridis',
    hue="Store"
)
plt.title("Sales Distribution by Store", fontsize=16)
plt.xlabel("Store", fontsize=12)
plt.ylabel("Sales", fontsize=12)
plt.grid(visible=True, linestyle="--", alpha=0.7)

# Wyświetlenie wykresu
plt.tight_layout()
plt.show()
```

```python

np.random.seed(42)
data = pd.DataFrame({
    "Store": np.random.choice(["Store_A", "Store_B", "Store_C", "Store_D"], size=24),
    "Month": np.tile(["Jan", "Feb", "Mar", "Apr", "May", "Jun"], 4),
    "Sales": np.random.randint(2000, 5000, size=24),
    "Region": np.random.choice(["North", "South", "East", "West"], size=24),
})


store_a = data[data["Store"] == "Store_A"].groupby("Month")["Sales"].sum()
store_b = data[data["Store"] == "Store_B"].groupby("Month")["Sales"].sum()

# Łączenie danych w jeden DataFrame
merged_data = pd.DataFrame({
    "Store_A_Sales": store_a,
    "Store_B_Sales": store_b
}).reset_index()  # Resetowanie indeksu, aby 'Month' była kolumną

# Usuwanie brakujących danych (jeśli jeden sklep nie miał sprzedaży w danym miesiącu)
merged_data = merged_data.dropna()


plt.figure(figsize=(8, 6))
sns.scatterplot(
    x="Store_A_Sales",
    y="Store_B_Sales",
    data=merged_data,
    color="blue",
    alpha=0.7,
    s=100  # Rozmiar punktów
)

# Dodanie tytułu, etykiet osi i siatki
plt.title("Scatter Plot: Sales Comparison (Store_A vs Store_B)", fontsize=16)
plt.xlabel("Store_A Sales", fontsize=12)
plt.ylabel("Store_B Sales", fontsize=12)
plt.grid(visible=True, linestyle="--", alpha=0.7)


plt.tight_layout()
plt.show()
```
