```python
import networkx as nx
import matplotlib.pyplot as plt

# Tworzenie pustego grafu
G = nx.Graph()

# Dodawanie węzłów
G.add_nodes_from(["Alice", "Bob", "Charlie", "David"])

# Dodawanie krawędzi (relacji)
G.add_edges_from([("Alice", "Bob"), ("Alice", "Charlie"), ("Bob", "David"), ("Charlie", "David")])

# Rysowanie grafu
plt.figure(figsize=(5,5))
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=12)
plt.show()
```

```python
print(dict(G.degree()))
```

```python
print(nx.shortest_path(G, source="Alice", target="David"))
```

```python
clustering_coefficients = nx.clustering(G)
print(clustering_coefficients)  # Współczynniki klasteryzacji dla każdego węzła

avg_clustering = nx.average_clustering(G)
print(f"Średni współczynnik klasteryzacji: {avg_clustering}")
```

```python
G_large = nx.erdos_renyi_graph(1000, 0.01)  # Symulacja losowego grafu 1000 węzłów

print(f"Liczba węzłów: {G_large.number_of_nodes()}")
print(f"Liczba krawędzi: {G_large.number_of_edges()}")
```
